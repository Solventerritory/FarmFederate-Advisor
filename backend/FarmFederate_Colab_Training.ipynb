```
import json, pandas as pd
from pathlib import Path

with open('../results/all_results.json') as f:
    results = json.load(f)
data = [{'Model': r['config']['name'], 'Type': r['config']['model_type'].upper(), 'F1': r['final_metrics']['f1_macro']} for r in results]
df = pd.DataFrame(data).sort_values('F1', ascending=False)
print(df.head(10))
print('Category Averages:')
print(df.groupby('Type')['F1'].agg(['mean', 'std']))
```
